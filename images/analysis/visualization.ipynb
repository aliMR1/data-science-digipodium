```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
penguins=sns.load_dataset("penguins")
penguins
```

```python
penguins['bill_length_mm'].plot()
```

```python
penguins['bill_length_mm'].plot(linestyle='',marker='s',markersize=7,alpha=0.6,figsize=(15,7),markerfacecolor='green',markeredgecolor='red')
```

```python
penguins['bill_length_mm'][:50].plot(kind='bar',figsize=(15,3),color='red')
```

```python
penguins.plot(kind='scatter',x='bill_length_mm',y='bill_depth_mm',s=45,alpha=0.6,marker='s')
```

subplots

```python
fig,axs=plt.subplots(1,2,figsize=(15,4),constrained_layout=True)
penguins['bill_length_mm'].plot(linestyle='',marker='s',markersize=7,alpha=0.6,figsize=(15,7),markerfacecolor='green',markeredgecolor='red',ax=axs[0])
penguins['bill_length_mm'][:50].plot(kind='bar',figsize=(15,3),color='red',ax=axs[1])
plt.show()
```

```python
penguins['species'].value_counts()
```

```python
penguins['species'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(12,5),title='species',ylabel='',wedgeprops=dict(width=0.7))
```
