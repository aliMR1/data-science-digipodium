```python
canada.rename(mapper={'AreaName':'Continent','RegName':'Region','DevName':'Status'},axis=1,inplace=True)
```

```python
updated_list=['Status','Continent','Region','Total']+years
print(updated_list)
```

grouping

```python
canada.groupby('Status').get_group('Developed regions')
```

```python
canada.groupby('Continent').get_group('Asia')
```

```python
canada.groupby('Status')[[2013,2010]].sum()
```

```python
canada.groupby('Status')[[2013,2010]].sum().plot(kind='pie',labels=['Developed','Still Developing'],subplots=True,autopct='%1.1f%%',figsize=(5,10),layout=(2,1))
```

```python
canada.groupby('Continent')['Total'].mean()
```

```python
canada.groupby('Continent')['Total'].agg(['mean','median'])
```

```python
canada.groupby(['Continent','Region'])['Total'].sum().reset_index()
```
